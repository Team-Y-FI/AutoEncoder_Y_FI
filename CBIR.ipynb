```python
import os, zipfile
import random
import numpy as np
import glob
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, random_split, Dataset
from torchvision import transforms
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
from PIL import Image
import torch.nn.functional as F
import hashlib
from dataclasses import dataclass
from typing import List, Tuple, Optional, Dict, Any

def seed_everything(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.benchmark = True

seed_everything(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("device:", device)

ZIP_PATH = "/content/drive/MyDrive/rang/11. 멀티모달/data/CBIR.zip"
EXTRACT_DIR = "/content/cbir_dataset"

assert os.path.exists(ZIP_PATH), f"zip 파일이 없습니다: {ZIP_PATH}"

os.makedirs(EXTRACT_DIR, exist_ok=True)
with zipfile.ZipFile(ZIP_PATH, 'r') as z:
    z.extractall(EXTRACT_DIR)

print("✅ extracted to:", EXTRACT_DIR)

for root, dirs, files in os.walk(EXTRACT_DIR):
    print(root, "dirs:", len(dirs), "files:", len(files))
    break

data_root = "/content/cbir_dataset/dataset"
assert os.path.isdir(data_root), f"data_root 폴더가 없습니다: {data_root}"

img_cnt = 0
for ext in ["*.jpg","*.jpeg","*.png","*.bmp","*.webp"]:
    img_cnt += len(glob.glob(os.path.join(data_root, ext)))
print("✅ images in data_root:", img_cnt)
assert img_cnt > 0, "data_root에 이미지가 없습니다. 경로를 다시 확인하세요."

IMG_SIZE = 128
BATCH_SIZE = 128
NUM_WORKERS = 2

train_tf = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
])

class AllImagesDataset(Dataset):
    def __init__(self, root, transform=None):
        self.root = root
        self.transform = transform

        exts = ["*.jpg","*.jpeg","*.png","*.bmp","*.webp"]
        paths = []
        for e in exts:
            paths += glob.glob(os.path.join(root, "**", e), recursive=True)
        self.paths = sorted(paths)

        assert len(self.paths) > 0, f"이미지를 못 찾았습니다. root={root}"

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        path = self.paths[idx]
        img = Image.open(path).convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img, path

dataset = AllImagesDataset(root=data_root, transform=train_tf)
print("num_images:", len(dataset))
print("sample_path:", dataset[0][1])

val_ratio = 0.1
val_size = int(len(dataset) * val_ratio)
train_size = len(dataset) - val_size

train_ds, val_ds = random_split(dataset, [train_size, val_size],
                                generator=torch.Generator().manual_seed(42))

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,
                          num_workers=NUM_WORKERS, pin_memory=True, drop_last=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=NUM_WORKERS, pin_memory=True, drop_last=False)

class ConvEncoder(nn.Module):
    def __init__(self, in_ch=3, base_ch=32, latent_ch=128):
        super().__init__()
        # 입력: (B, 3, 128, 128)
        self.net = nn.Sequential(
            # -> (B, 32, 64, 64)
            nn.Conv2d(in_ch, base_ch, kernel_size=3, stride=2, padding=1),
            nn.ReLU(inplace=True),

            # -> (B, 64, 32, 32)
            nn.Conv2d(base_ch, base_ch*2, kernel_size=3, stride=2, padding=1),
            nn.ReLU(inplace=True),

            # -> (B, 128, 16, 16)
            nn.Conv2d(base_ch*2, base_ch*4, kernel_size=3, stride=2, padding=1),
            nn.ReLU(inplace=True),

            # -> (B, latent_ch, 8, 8)
            nn.Conv2d(base_ch*4, latent_ch, kernel_size=3, stride=2, padding=1),
            nn.ReLU(inplace=True),
        )

    def forward(self, x):
        return self.net(x)

class ConvDecoder(nn.Module):
    def __init__(self, out_ch=3, base_ch=32, latent_ch=128):
        super().__init__()
        # 입력: (B, latent_ch, 8, 8)
        self.net = nn.Sequential(
            # -> (B, 128, 16, 16)
            nn.ConvTranspose2d(latent_ch, base_ch*4, kernel_size=4, stride=2, padding=1),
            nn.ReLU(inplace=True),

            # -> (B, 64, 32, 32)
            nn.ConvTranspose2d(base_ch*4, base_ch*2, kernel_size=4, stride=2, padding=1),
            nn.ReLU(inplace=True),

            # -> (B, 32, 64, 64)
            nn.ConvTranspose2d(base_ch*2, base_ch, kernel_size=4, stride=2, padding=1),
            nn.ReLU(inplace=True),

            # -> (B, 3, 128, 128)
            nn.ConvTranspose2d(base_ch, out_ch, kernel_size=4, stride=2, padding=1),
            nn.Sigmoid(),
        )

    def forward(self, z):
        return self.net(z)

class ConvAutoencoder(nn.Module):
    def __init__(self, in_ch=3, base_ch=32, latent_ch=128):
        super().__init__()
        self.encoder = ConvEncoder(in_ch, base_ch, latent_ch)
        self.decoder = ConvDecoder(in_ch, base_ch, latent_ch)

    def forward(self, x):
        z = self.encoder(x)
        x_hat = self.decoder(z)
        return x_hat, z

model = ConvAutoencoder(in_ch=3, base_ch=32, latent_ch=128).to(device)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)

use_amp = (device.type == "cuda")
scaler = torch.cuda.amp.GradScaler(enabled=use_amp)

@torch.no_grad()
def evaluate(model, loader, criterion):
    model.eval()
    total_loss, n = 0.0, 0
    for x, _ in loader:
        x = x.to(device, non_blocking=True)
        x_hat, _ = model(x)
        loss = criterion(x_hat, x)
        total_loss += loss.item() * x.size(0)
        n += x.size(0)
    return total_loss / max(n, 1)

def train_one_epoch(model, loader, optimizer, criterion, scaler):
    model.train()
    total_loss, n = 0.0, 0

    for x, _ in tqdm(loader, desc="train", leave=False):
        x = x.to(device, non_blocking=True)
        optimizer.zero_grad(set_to_none=True)

        with torch.cuda.amp.autocast(enabled=use_amp):
            x_hat, _ = model(x)
            loss = criterion(x_hat, x)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        total_loss += loss.item() * x.size(0)
        n += x.size(0)

    return total_loss / max(n, 1)

EPOCHS = 20
best_val = float("inf")
save_path = "/content/conv_autoencoder_best.pt"

history = {"train_loss": [], "val_loss": []}

for epoch in range(1, EPOCHS + 1):
    train_loss = train_one_epoch(model, train_loader, optimizer, criterion, scaler)
    val_loss = evaluate(model, val_loader, criterion)

    history["train_loss"].append(train_loss)
    history["val_loss"].append(val_loss)

    print(f"[Epoch {epoch:02d}/{EPOCHS}] train_loss={train_loss:.6f} | val_loss={val_loss:.6f}")

    if val_loss < best_val:
        best_val = val_loss
        torch.save({
            "model_state_dict": model.state_dict(),
            "img_size": IMG_SIZE,
            "base_ch": 32,
            "latent_ch": 128,
        }, save_path)
        print(f"✅ best saved: {save_path} (val_loss={best_val:.6f})")

@torch.no_grad()
def show_reconstructions(model, loader, n_images=8):
    model.eval()
    x, _ = next(iter(loader))
    x = x[:n_images].to(device)
    x_hat, _ = model(x)

    grid_in  = make_grid(x.cpu(), nrow=n_images, padding=2)
    grid_out = make_grid(x_hat.cpu(), nrow=n_images, padding=2)

    plt.figure(figsize=(18, 4))
    plt.title("Original")
    plt.imshow(np.transpose(grid_in.numpy(), (1, 2, 0)))
    plt.axis("off")
    plt.show()

    plt.figure(figsize=(18, 4))
    plt.title("Reconstructed")
    plt.imshow(np.transpose(grid_out.numpy(), (1, 2, 0)))
    plt.axis("off")
    plt.show()

show_reconstructions(model, val_loader, n_images=8)

plt.figure(figsize=(6,4))
plt.plot(history["train_loss"], label="train")
plt.plot(history["val_loss"], label="val")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
CKPT_PATH = "/content/conv_autoencoder_best.pt"
DATA_ROOT = "/content/cbir_dataset/dataset"
MODEL_CLASS = ConvAutoencoder

SAVE_DIR = "/content/artifacts"
os.makedirs(SAVE_DIR, exist_ok=True)

ckpt = torch.load(CKPT_PATH, map_location="cpu")

img_size  = ckpt["img_size"]
base_ch   = ckpt["base_ch"]
latent_ch = ckpt["latent_ch"]

model = MODEL_CLASS(in_ch=3, base_ch=base_ch, latent_ch=latent_ch).to(device)
model.load_state_dict(ckpt["model_state_dict"])
model.eval()

print("✅ loaded ckpt:", CKPT_PATH)
print("   img_size:", img_size, "base_ch:", base_ch, "latent_ch:", latent_ch)

dataset_all = AllImagesDataset(root=DATA_ROOT, transform=train_tf)
loader_all = DataLoader(dataset_all, batch_size=128, shuffle=False, num_workers=2, pin_memory=True)

@torch.no_grad()
def extract_latents(model, loader, device, l2_normalize=True):
    feats_list = []
    paths_list = []

    for x, paths in tqdm(loader, desc="extract", leave=False):
        x = x.to(device, non_blocking=True)
        z = model.encoder(x)
        z = z.flatten(1)

        if l2_normalize:
            z = torch.nn.functional.normalize(z, p=2, dim=1)

        feats_list.append(z.cpu())
        paths_list.extend(list(paths))

    feats = torch.cat(feats_list, dim=0).numpy()
    return feats, paths_list

features, paths = extract_latents(model, loader_all, device, l2_normalize=True)

print("✅ features shape:", features.shape)
print("✅ example path:", paths[0])

np.savez_compressed(
    os.path.join(SAVE_DIR, "latent_features.npz"),
    features=features,
    paths=np.array(paths, dtype=object),
)

torch.save(
    {"features": torch.from_numpy(features), "paths": paths},
    os.path.join(SAVE_DIR, "latent_features.pt"),
)

print("✅ saved:", os.path.join(SAVE_DIR, "latent_features.npz"))
print("✅ saved:", os.path.join(SAVE_DIR, "latent_features.pt"))
```

```python
# =========================================
# Top-K Retrieval (Cosine / L2) + Grid 시각화
# =========================================
CKPT_PATH = "/content/conv_autoencoder_best.pt"
FEATURES_NPZ = "/content/artifacts/latent_features.npz"
DATA_ROOT = "/content/cbir_dataset/dataset"
IMG_SIZE = 128

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def load_latent_db(npz_path: str):
    assert os.path.exists(npz_path), f"NPZ 파일이 없습니다: {npz_path}"

    db = np.load(npz_path, allow_pickle=True)
    keys = list(db.keys())

    path_key_candidates = ["paths", "path", "files", "filenames", "image_paths", "img_paths"]
    feat_key_candidates = ["features", "feats", "latent", "latent_features", "embeddings", "z"]

    path_key = None
    feat_key = None

    for k in path_key_candidates:
        if k in keys:
            path_key = k
            break
    for k in feat_key_candidates:
        if k in keys:
            feat_key = k
            break

    if path_key is None:
        for k in keys:
            arr = db[k]
            if arr.ndim == 1 and len(arr) > 0 and (isinstance(arr[0], (str, bytes, np.str_, np.bytes_))):
                path_key = k
                break

    if feat_key is None:
        for k in keys:
            arr = db[k]
            if arr.ndim == 2:
                feat_key = k
                break

    assert path_key is not None, f"paths에 해당하는 키를 못 찾았습니다. npz keys={keys}"
    assert feat_key is not None, f"features에 해당하는 키를 못 찾았습니다. npz keys={keys}"

    paths = db[path_key]
    feats = db[feat_key].astype(np.float32)

    paths_clean = []
    for p in paths:
        if isinstance(p, (bytes, np.bytes_)):
            paths_clean.append(p.decode("utf-8"))
        else:
            paths_clean.append(str(p))

    assert feats.ndim == 2, f"features shape이 2D가 아닙니다: {feats.shape}"
    assert len(paths_clean) == feats.shape[0], f"paths 길이({len(paths_clean)})와 feats N({feats.shape[0]}) 불일치"

    return paths_clean, feats, {"path_key": path_key, "feat_key": feat_key, "keys": keys}

def load_image_rgb(path: str, img_size: int):
    img = Image.open(path).convert("RGB")
    img = img.resize((img_size, img_size))
    return img


def pil_to_tensor(img: Image.Image):
    arr = np.array(img).astype(np.float32) / 255.0
    arr = np.transpose(arr, (2, 0, 1))
    t = torch.from_numpy(arr).unsqueeze(0)
    return t

def get_query_feature_from_db(query_path: str, db_paths: list, db_feats: np.ndarray):
    if query_path in db_paths:
        idx = db_paths.index(query_path)
        return db_feats[idx], idx

    q_base = os.path.basename(query_path)
    base_map = {os.path.basename(p): i for i, p in enumerate(db_paths)}
    if q_base in base_map:
        idx = base_map[q_base]
        return db_feats[idx], idx

    return None, None


def try_encode_query_with_ckpt(query_path: str, ckpt_path: str):
    assert os.path.exists(ckpt_path), f"CKPT가 없습니다: {ckpt_path}"

    ckpt = torch.load(ckpt_path, map_location=DEVICE)

    if isinstance(ckpt, torch.nn.Module):
        model = ckpt.to(DEVICE).eval()
        img_size = IMG_SIZE
    elif isinstance(ckpt, dict):
        base_ch   = ckpt.get("base_ch", 32)
        latent_ch = ckpt.get("latent_ch", 128)
        img_size  = ckpt.get("img_size", IMG_SIZE)

        model = ConvAutoencoder(in_ch=3, base_ch=base_ch, latent_ch=latent_ch).to(DEVICE)
        state_dict = ckpt.get("model_state_dict", None)
        if state_dict is None:
            if all(isinstance(v, torch.Tensor) for v in ckpt.values()):
                state_dict = ckpt
            else:
                raise RuntimeError("ckpt에 model_state_dict가 없습니다.")
        model.load_state_dict(state_dict, strict=True)
        model.eval()
    else:
        raise RuntimeError("지원하지 않는 ckpt 포맷입니다.")

    tf = transforms.Compose([
        transforms.Resize((img_size, img_size)),
        transforms.ToTensor(),
    ])

    img = Image.open(query_path).convert("RGB")
    x = tf(img).unsqueeze(0).to(DEVICE)

    with torch.no_grad():
        z = model.encoder(x)
        z = z.flatten(1)
        z = F.normalize(z, p=2, dim=1)
        feat = z.squeeze(0).cpu().numpy()

    return feat

def topk_search(db_feats: np.ndarray, query_feat: np.ndarray, k: int = 8, metric: str = "cosine", exclude_index: int = None):
    db = torch.from_numpy(db_feats).to(DEVICE)
    q = torch.from_numpy(query_feat).to(DEVICE).unsqueeze(0)

    if metric.lower() == "cosine":
        db_n = F.normalize(db, dim=1)
        q_n = F.normalize(q, dim=1)
        scores = (db_n @ q_n.t()).squeeze(1)

        if exclude_index is not None:
            scores[exclude_index] = -1e9
        topk_vals, topk_idx = torch.topk(scores, k=k, largest=True)
        return topk_idx.detach().cpu().numpy(), topk_vals.detach().cpu().numpy()

    elif metric.lower() in ["l2", "euclidean"]:
        dists = torch.norm(db - q, p=2, dim=1)
        if exclude_index is not None:
            dists[exclude_index] = 1e9
        topk_vals, topk_idx = torch.topk(dists, k=k, largest=False)
        return topk_idx.detach().cpu().numpy(), topk_vals.detach().cpu().numpy()

    else:
        raise ValueError(f"지원하지 않는 metric: {metric}")

def show_retrieval_grid(query_path: str, retrieved_paths: list, retrieved_scores: np.ndarray, metric: str, img_size: int, ncols: int = 5):
    k = len(retrieved_paths)
    nrows = int(np.ceil((k + 1) / ncols))

    plt.figure(figsize=(ncols * 3, nrows * 3))

    ax = plt.subplot(nrows, ncols, 1)
    q_img = load_image_rgb(query_path, img_size)
    ax.imshow(q_img)
    ax.set_title("Query", fontsize=12)
    ax.axis("off")

    for i, (p, s) in enumerate(zip(retrieved_paths, retrieved_scores), start=2):
        ax = plt.subplot(nrows, ncols, i)
        img = load_image_rgb(p, img_size)
        ax.imshow(img)

        if metric.lower() == "cosine":
            ax.set_title(f"Top{i-1}\ncos={float(s):.4f}", fontsize=10)
        else:
            ax.set_title(f"Top{i-1}\nL2={float(s):.4f}", fontsize=10)

        ax.axis("off")

    plt.tight_layout()
    plt.show()

db_paths, db_feats, meta = load_latent_db(FEATURES_NPZ)
print(f"[DB] loaded: N={len(db_paths)}, D={db_feats.shape[1]}")
print(f"[DB] npz keys={meta['keys']}")
print(f"[DB] using path_key='{meta['path_key']}', feat_key='{meta['feat_key']}'")

QUERY_PATH = os.path.join(DATA_ROOT, db_paths[0]) if not os.path.isabs(db_paths[0]) else db_paths[0]

TOP_K = 8
EXCLUDE_SELF = True

q_feat, q_idx = get_query_feature_from_db(QUERY_PATH, db_paths, db_feats)

if q_feat is not None:
    print(f"[Query] DB에서 feature 재사용 성공 (idx={q_idx})")
else:
    print("[Query] DB에 동일 path를 못 찾아서 CKPT로 feature 추출을 시도합니다.")
    q_feat = try_encode_query_with_ckpt(
        query_path=QUERY_PATH,
        ckpt_path=CKPT_PATH,
        latent_dim=db_feats.shape[1],
        img_size=IMG_SIZE,
    )
    q_idx = None
    print("[Query] CKPT로 feature 추출 성공")

exclude_index = q_idx if (EXCLUDE_SELF and q_idx is not None) else None

cos_idx, cos_scores = topk_search(db_feats, q_feat, k=TOP_K, metric="cosine", exclude_index=exclude_index)
cos_paths = [db_paths[i] for i in cos_idx]
cos_paths = [p if os.path.isabs(p) else os.path.join(DATA_ROOT, p) for p in cos_paths]

print("\n[Cosine Top-K]")
for rank, (p, s) in enumerate(zip(cos_paths, cos_scores), start=1):
    print(f"{rank:02d}. score={float(s):.6f}  path={p}")

show_retrieval_grid(QUERY_PATH, cos_paths, cos_scores, metric="cosine", img_size=IMG_SIZE, ncols=5)

l2_idx, l2_scores = topk_search(db_feats, q_feat, k=TOP_K, metric="l2", exclude_index=exclude_index)
l2_paths = [db_paths[i] for i in l2_idx]
l2_paths = [p if os.path.isabs(p) else os.path.join(DATA_ROOT, p) for p in l2_paths]

print("\n[L2 Top-K]")
for rank, (p, s) in enumerate(zip(l2_paths, l2_scores), start=1):
    print(f"{rank:02d}. dist={float(s):.6f}  path={p}")

show_retrieval_grid(QUERY_PATH, l2_paths, l2_scores, metric="l2", img_size=IMG_SIZE, ncols=5)
```

```python
def _safe_basename(p: str) -> str:
    return os.path.basename(str(p).rstrip("/\\ "))

def _is_abs_path(p: str) -> bool:
    return os.path.isabs(str(p))

def _real(p: str) -> str:
    try:
        return os.path.realpath(p)
    except Exception:
        return p

def _exists(p: str) -> bool:
    try:
        return os.path.exists(p)
    except Exception:
        return False

def _file_size(p: str) -> Optional[int]:
    try:
        return os.path.getsize(p)
    except Exception:
        return None

def _quick_md5(p: str, max_bytes: int = 1024 * 1024) -> Optional[str]:
    try:
        h = hashlib.md5()
        with open(p, "rb") as f:
            h.update(f.read(max_bytes))
        return h.hexdigest()
    except Exception:
        return None

def resolve_paths_to_disk(npz_paths: List[str], data_root: str) -> List[str]:
    resolved = []
    for p in npz_paths:
        p = str(p)

        if _is_abs_path(p) and _exists(p):
            resolved.append(p)
            continue

        cand1 = os.path.join(data_root, p)
        if _exists(cand1):
            resolved.append(cand1)
            continue

        cand2 = os.path.join(data_root, _safe_basename(p))
        if _exists(cand2):
            resolved.append(cand2)
            continue

        resolved.append(cand1)

    return resolved

def normalize_query_path(query_path: str, data_root: str) -> str:
    query_path = str(query_path)
    if _is_abs_path(query_path):
        return query_path
    return os.path.join(data_root, query_path)

def _format_tile_title(rank: int, file_path: str, value: float, metric: str) -> str:
    fname = _safe_basename(file_path)
    metric = metric.lower()
    if metric in ("cosine", "dot"):
        return f"{rank}) {fname}\n{metric}={value:.4f}"
    return f"{rank}) {fname}\n{metric}={value:.4f}"

@dataclass
class RetrievalResult:
    indices: np.ndarray
    values: np.ndarray
    metric: str
    is_similarity: bool

@dataclass
class VisualizeConfig:
    img_size: int = 128
    cols: int = 6
    figsize_per_tile: float = 2.6

class LatentFeatureIndex:
    def __init__(self, features_npz: str, data_root: str):
        self.features_npz = features_npz
        self.data_root = data_root

        loaded = np.load(features_npz, allow_pickle=True)
        if "paths" not in loaded or "features" not in loaded:
            raise ValueError(f"NPZ에 'paths'와 'features' 키가 필요합니다: {features_npz}")

        npz_paths = loaded["paths"]
        feats = loaded["features"]

        self.npz_paths = [str(p) for p in (npz_paths.tolist() if hasattr(npz_paths, "tolist") else npz_paths)]
        self.disk_paths = resolve_paths_to_disk(self.npz_paths, data_root)

        if not isinstance(feats, np.ndarray) or feats.ndim != 2:
            raise ValueError(f"'features'는 (N, D) 2D numpy array 여야 합니다. got shape={getattr(feats,'shape',None)}")
        self.features = feats.astype(np.float32, copy=False)
        self.N, self.D = self.features.shape

        self.basename_to_indices: Dict[str, List[int]] = {}
        for i, p in enumerate(self.npz_paths):
            b = _safe_basename(p)
            self.basename_to_indices.setdefault(b, []).append(i)

        self.realpath_to_index: Dict[str, int] = {}
        for i, p in enumerate(self.disk_paths):
            if _exists(p):
                self.realpath_to_index[_real(p)] = i

        self.basename_size_to_indices: Dict[Tuple[str, int], List[int]] = {}
        for i, p in enumerate(self.disk_paths):
            if not _exists(p):
                continue
            key = (_safe_basename(p), int(os.path.getsize(p)))
            self.basename_size_to_indices.setdefault(key, []).append(i)

    def _resolve_query_index(self, query_path: str) -> int:
        q_in = str(query_path)
        q_abs = normalize_query_path(q_in, self.data_root)

        q_disk = q_abs if _exists(q_abs) else (q_in if _exists(q_in) else None)

        candidates = set([q_in, q_abs])
        if q_disk is not None:
            candidates.add(_real(q_disk))
        try:
            rel = os.path.relpath(q_abs, self.data_root)
            candidates.add(rel)
        except Exception:
            pass

        for i, p in enumerate(self.npz_paths):
            if p in candidates:
                return i
        for i, p in enumerate(self.disk_paths):
            if p in candidates:
                return i
            if _exists(p) and _real(p) in candidates:
                return i

        q_base = _safe_basename(q_in)
        if q_base in self.basename_to_indices:
            return self.basename_to_indices[q_base][0]
        if q_disk is not None:
            q_real = _real(q_disk)
            if q_real in self.realpath_to_index:
                return self.realpath_to_index[q_real]

            qs = _file_size(q_disk)
            if qs is not None:
                key = (q_base, int(qs))
                cand_idxs = self.basename_size_to_indices.get(key, [])
                if len(cand_idxs) == 1:
                    return cand_idxs[0]
                elif len(cand_idxs) > 1:
                    # 3-C) quick md5로 확정
                    qh = _quick_md5(q_disk)
                    if qh is not None:
                        for ci in cand_idxs:
                            ph = _quick_md5(self.disk_paths[ci])
                            if ph == qh:
                                return ci

        sample = "\n".join([f"- {self.npz_paths[i]}" for i in range(min(12, len(self.npz_paths)))])
        hint_candidates = ""
        if q_base:
            found = []
            try:
                for name in os.listdir(self.data_root):
                    p = os.path.join(self.data_root, name)
                    if os.path.isfile(p) and _safe_basename(p) == q_base:
                        found.append(p)
                if found:
                    hint_candidates = "\n".join([f"- {p}" for p in found[:10]])
            except Exception:
                pass

        msg = (
            f"[Query 매칭 실패]\n"
            f"입력 query_path: {q_in}\n"
            f"DATA_ROOT 적용 경로: {q_abs}\n"
            f"basename: {q_base}\n\n"
            f"현재 NPZ paths가 예시처럼 DATA_ROOT 루트에 평탄화(flat)된 파일명(0.jpg, 1.jpg, ...)로 보입니다.\n"
            f"따라서 query_path도 '0.jpg' 같은 형태로 주는 것이 가장 확실합니다.\n\n"
            f"NPZ paths 예시:\n{sample}\n"
        )
        if hint_candidates:
            msg += f"\nDATA_ROOT에서 동일 basename 후보:\n{hint_candidates}\n"
        msg += (
            "\n해결 팁:\n"
            "1) query_path를 '0.jpg' 처럼 파일명만으로 넣어보세요.\n"
            "2) 또는 NPZ에서 아무 path 하나를 골라 그대로 query_path로 넣어보세요.\n"
        )
        raise FileNotFoundError(msg)

    def search(self, query_path: str, top_k: int = 10, metric: str = "cosine", include_query: bool = True) -> RetrievalResult:
        metric = metric.lower().strip()
        if top_k <= 0:
            raise ValueError("top_k는 1 이상이어야 합니다.")

        q_idx = self._resolve_query_index(query_path)
        q = self.features[q_idx]
        X = self.features

        if metric in ("cosine", "dot"):
            vals = X @ q
            is_similarity = True
            order = np.argpartition(-vals, kth=min(top_k + 1, self.N - 1))[: min(top_k + 1, self.N)]
            order = order[np.argsort(-vals[order])]
        elif metric == "l2":
            dif = X - q[None, :]
            vals = np.sqrt(np.sum(dif * dif, axis=1))
            is_similarity = False
            order = np.argpartition(vals, kth=min(top_k + 1, self.N - 1))[: min(top_k + 1, self.N)]
            order = order[np.argsort(vals[order])]
        elif metric == "l1":
            vals = np.sum(np.abs(X - q[None, :]), axis=1)
            is_similarity = False
            order = np.argpartition(vals, kth=min(top_k + 1, self.N - 1))[: min(top_k + 1, self.N)]
            order = order[np.argsort(vals[order])]
        else:
            raise ValueError(f"지원하지 않는 metric: {metric}. 지원: cosine, dot, l2, l1")

        order_list = order.tolist()
        if include_query:
            order_list = [i for i in order_list if i != q_idx]
            order_list = [q_idx] + order_list

        order_list = order_list[:top_k]
        out_idx = np.array(order_list, dtype=np.int64)
        out_vals = vals[out_idx].astype(np.float32)

        return RetrievalResult(indices=out_idx, values=out_vals, metric=metric, is_similarity=is_similarity)

def _load_image_for_imshow(path: str) -> np.ndarray:
    try:
        return plt.imread(path)
    except Exception:
        return np.zeros((64, 64, 3), dtype=np.float32)

def plot_retrieval_grid(
    disk_paths: List[str],
    result: RetrievalResult,
    cfg: VisualizeConfig,
    title_prefix: str = "",
    save_path: Optional[str] = None,
) -> plt.Figure:
    K = len(result.indices)
    cols = max(1, int(cfg.cols))
    rows = int(np.ceil(K / cols))

    fig = plt.figure(figsize=(cols * cfg.figsize_per_tile, rows * cfg.figsize_per_tile))

    for r in range(K):
        idx = int(result.indices[r])
        img_path = disk_paths[idx]
        img = _load_image_for_imshow(img_path)

        ax = fig.add_subplot(rows, cols, r + 1)
        ax.imshow(img)
        ax.axis("off")

        val = float(result.values[r])
        tile_title = _format_tile_title(rank=r, file_path=img_path, value=val, metric=result.metric)
        if r == 0:
            ax.set_title(f"[Query]\n{tile_title}", fontsize=9)
        else:
            ax.set_title(tile_title, fontsize=9)

    if title_prefix:
        fig.suptitle(f"{title_prefix} (metric={result.metric})", fontsize=14)
    else:
        fig.suptitle(f"Retrieval Results (metric={result.metric})", fontsize=14)

    fig.tight_layout()
    fig.subplots_adjust(top=0.90)

    if save_path is not None:
        if os.path.dirname(save_path):
            os.makedirs(os.path.dirname(save_path), exist_ok=True)
        fig.savefig(save_path, dpi=150)

    return fig

def visualize_cbir_results(
    features_npz: str,
    data_root: str,
    query_path: str,
    top_k: int = 12,
    mode: str = "single",
    metric: str = "cosine",
    compare_metrics: Tuple[str, str] = ("cosine", "l1"),
    cols: int = 6,
    save_path: Optional[str] = None,
    img_size: int = 128,
) -> Dict[str, Any]:
    index = LatentFeatureIndex(features_npz=features_npz, data_root=data_root)
    cfg = VisualizeConfig(img_size=img_size, cols=cols)

    mode = mode.lower().strip()
    out: Dict[str, Any] = {"mode": mode, "query_path": query_path, "top_k": top_k}

    if mode == "single":
        res = index.search(query_path=query_path, top_k=top_k, metric=metric, include_query=True)
        fig = plot_retrieval_grid(index.disk_paths, res, cfg, title_prefix="CBIR Top-K", save_path=save_path)
        out.update({"result": res, "figure": fig, "save_path": save_path})
        return out

    if mode == "compare":
        m1, m2 = compare_metrics
        res1 = index.search(query_path=query_path, top_k=top_k, metric=m1, include_query=True)
        res2 = index.search(query_path=query_path, top_k=top_k, metric=m2, include_query=True)

        K = top_k
        cols = max(1, int(cols))
        rows = int(np.ceil(K / cols))

        fig = plt.figure(figsize=(cols * 2 * cfg.figsize_per_tile, rows * cfg.figsize_per_tile))

        for r in range(K):
            idx = int(res1.indices[r])
            img_path = index.disk_paths[idx]
            img = _load_image_for_imshow(img_path)

            ax = fig.add_subplot(rows, cols * 2, r * 2 + 1)
            ax.imshow(img)
            ax.axis("off")

            val = float(res1.values[r])
            tile_title = _format_tile_title(rank=r, file_path=img_path, value=val, metric=res1.metric)
            ax.set_title(f"[Query]\n{tile_title}" if r == 0 else tile_title, fontsize=9)

        for r in range(K):
            idx = int(res2.indices[r])
            img_path = index.disk_paths[idx]
            img = _load_image_for_imshow(img_path)

            ax = fig.add_subplot(rows, cols * 2, r * 2 + 2)
            ax.imshow(img)
            ax.axis("off")

            val = float(res2.values[r])
            tile_title = _format_tile_title(rank=r, file_path=img_path, value=val, metric=res2.metric)
            ax.set_title(f"[Query]\n{tile_title}" if r == 0 else tile_title, fontsize=9)

        fig.suptitle(f"CBIR Compare: {m1} (left) vs {m2} (right)", fontsize=14)
        fig.tight_layout()
        fig.subplots_adjust(top=0.90)

        if save_path is not None:
            if os.path.dirname(save_path):
                os.makedirs(os.path.dirname(save_path), exist_ok=True)
            fig.savefig(save_path, dpi=150)

        out.update({"result_left": res1, "result_right": res2, "figure": fig, "save_path": save_path})
        return out

    raise ValueError(f"mode는 'single' 또는 'compare'만 지원합니다. got: {mode}")

FEATURES_NPZ = "/content/artifacts/latent_features.npz"
DATA_ROOT = "/content/cbir_dataset/dataset"
IMG_SIZE = 128

QUERY_PATH = "0.jpg"

_ = visualize_cbir_results(
    features_npz=FEATURES_NPZ,
    data_root=DATA_ROOT,
    query_path=QUERY_PATH,
    top_k=12,
    mode="single",
    metric="cosine",
    cols=6,
    save_path=None,
    img_size=IMG_SIZE,
)
plt.show()

_ = visualize_cbir_results(
    features_npz=FEATURES_NPZ,
    data_root=DATA_ROOT,
    query_path=QUERY_PATH,
    top_k=12,
    mode="compare",
    compare_metrics=("cosine", "l1"),
    cols=6,
    save_path=None,
    img_size=IMG_SIZE,
)
plt.show()
```

```python
# ============================================================
# 여러 Query를 한 번에 점검
# ============================================================
def pick_query_paths_from_npz(
    features_npz: str,
    num_queries: int = 8,
    seed: int = 42,
    prefer_unique_basenames: bool = True,
) -> List[str]:
    loaded = np.load(features_npz, allow_pickle=True)
    paths = loaded["paths"]
    paths = [str(p) for p in (paths.tolist() if hasattr(paths, "tolist") else paths)]

    rng = np.random.default_rng(seed)

    if not prefer_unique_basenames:
        idxs = rng.choice(len(paths), size=min(num_queries, len(paths)), replace=False)
        return [paths[i] for i in idxs.tolist()]

    seen = set()
    uniq = []
    for p in paths:
        b = os.path.basename(p)
        if b in seen:
            continue
        seen.add(b)
        uniq.append(p)

    base_pool = uniq if len(uniq) >= num_queries else paths
    idxs = rng.choice(len(base_pool), size=min(num_queries, len(base_pool)), replace=False)
    return [base_pool[i] for i in idxs.tolist()]


def batch_visualize_queries(
    features_npz: str,
    data_root: str,
    query_paths: Optional[List[str]] = None,
    num_queries: int = 8,
    seed: int = 42,
    top_k: int = 12,
    mode: str = "single",
    metric: str = "cosine",
    compare_metrics: Tuple[str, str] = ("cosine", "l1"),
    cols: int = 6,
    save_dir: Optional[str] = None,
    img_size: int = 128,
    prefer_unique_basenames: bool = True,
) -> List[Dict[str, Any]]:
    index = LatentFeatureIndex(features_npz=features_npz, data_root=data_root)

    if query_paths is None:
        query_paths = pick_query_paths_from_npz(
            features_npz=features_npz,
            num_queries=num_queries,
            seed=seed,
            prefer_unique_basenames=prefer_unique_basenames,
        )

    results = []
    mode = mode.lower().strip()

    if save_dir is not None:
        os.makedirs(save_dir, exist_ok=True)

    for qi, qp in enumerate(query_paths):
        base = os.path.basename(str(qp))
        print(f"\n===== [{qi+1}/{len(query_paths)}] Query: {qp} (basename={base}) =====")

        if mode == "single":
            res = index.search(query_path=qp, top_k=top_k, metric=metric, include_query=True)
            save_path = None
            if save_dir is not None:
                save_path = os.path.join(save_dir, f"{qi:02d}_{base}_{metric}.png")

            fig = plot_retrieval_grid(
                disk_paths=index.disk_paths,
                result=res,
                cfg=VisualizeConfig(img_size=img_size, cols=cols),
                title_prefix=f"CBIR Top-K | Query={base}",
                save_path=save_path,
            )
            plt.show()

            results.append({
                "query_path": qp,
                "metric": metric,
                "result": res,
                "save_path": save_path,
            })

        elif mode == "compare":
            m1, m2 = compare_metrics
            save_path = None
            if save_dir is not None:
                save_path = os.path.join(save_dir, f"{qi:02d}_{base}_{m1}_vs_{m2}.png")

            out = visualize_cbir_results(
                features_npz=features_npz,
                data_root=data_root,
                query_path=qp,
                top_k=top_k,
                mode="compare",
                compare_metrics=(m1, m2),
                cols=cols,
                save_path=save_path,
                img_size=img_size,
            )
            plt.show()

            results.append({
                "query_path": qp,
                "compare_metrics": (m1, m2),
                "result_left": out["result_left"],
                "result_right": out["result_right"],
                "save_path": save_path,
            })
        else:
            raise ValueError("mode는 'single' 또는 'compare'만 가능합니다.")

    return results


FEATURES_NPZ = "/content/artifacts/latent_features.npz"
DATA_ROOT = "/content/cbir_dataset/dataset"
IMG_SIZE = 128

_ = batch_visualize_queries(
    features_npz=FEATURES_NPZ,
    data_root=DATA_ROOT,
    query_paths=None,
    num_queries=8,
    seed=7,
    top_k=12,
    mode="single",
    metric="cosine",
    cols=6,
    save_dir=None,
    img_size=IMG_SIZE,
)
```
